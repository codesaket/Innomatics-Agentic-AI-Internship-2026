## Problem 1 — Social Media Post Engagement Analyzer
This function calculates total likes and determines whether the post is viral or has normal engagement.

```python
def analyze_engagement(likes):
    total_likes = sum(likes)
    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

likes = [200, 300, 150, 400]
analyze_engagement(likes)
```

## Problem 3 — Rainfall Adequacy Checker
This function calculates average rainfall and checks if it meets the required level.

```python
def rainfall_checker(rainfall, required):
    average = sum(rainfall) / len(rainfall)
    print("Average Rainfall:", average)

    if average >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

rainfall_data = [70, 75, 68, 80, 67]
rainfall_checker(rainfall_data, 70)
```

## Problem 4 — Duplicate Account Detection
This function checks if duplicate usernames exist using a set.

```python
def check_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["abhi", "rahul", "abhi", "neha"]
check_duplicates(users)
```

## Problem 6 — Premium Crop Price Filter
This function filters crop prices above ₹2000.

```python
def premium_crops(prices):
    premium = [price for price in prices if price > 2000]
    print("Premium Crops:", premium)

crop_prices = [1500, 2500, 1800, 3200]
premium_crops(crop_prices)
```

## Problem 7 — Application Health Checker
This function determines system health based on error count.

```python
def system_health(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

system_health(7)
```

## Problem 9 — Student Attendance Eligibility
This function calculates attendance percentage and checks eligibility.

```python
def attendance_checker(attendance):
    present = attendance.count("P")
    percentage = (present / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

attendance = ["P", "P", "P", "P", "A"]
attendance_checker(attendance)
```

## Problem 10 — Smart Electricity Bill Analyzer
This function calculates electricity bill based on slab rates and classifies usage.

```python
def electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


bill, status = electricity_bill(250)
print("Total Bill:", bill)
print("Usage Level:", status)
```
