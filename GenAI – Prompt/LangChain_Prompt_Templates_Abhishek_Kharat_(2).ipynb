## ✅ Task 1: Replace Hardcoded Prompts (10%)

**Goal:** Convert a hardcoded f-string function into a reusable `PromptTemplate`.

**Given (Bad) Code:**
```python
def explain_topic(topic):
    return f"Explain {topic} in simple terms for beginners"
```

**Rules:** No f-strings | Use `PromptTemplate` | Must be reusable

````python
# ─────────────────────────────────────────────────────────────
# TASK 1: Replace Hardcoded Prompt with LangChain PromptTemplate
# ─────────────────────────────────────────────────────────────

# Define reusable PromptTemplate — no f-strings, fully dynamic
explain_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

# Function that uses the template (logic kept separate from template)
def explain_topic(topic):
    """Generate a beginner-friendly explanation prompt for a given topic."""
    return explain_template.format(topic=topic)

# ── Test with multiple topics ──
test_topics = ["Machine Learning", "Neural Networks", "Python"]

print("📌 Task 1 Output – Reusable PromptTemplate:")
print("-" * 50)
for t in test_topics:
    print(explain_topic(t))
print("-" * 50)
````

## ✅ Task 2: Multi-Input Prompt System (15%)

**Template:** `Explain {topic} for {audience} in a {tone} tone`

**Test Cases:** AI|beginners|friendly · Python|kids|fun · Deep Learning|engineers|technical

````python
# ─────────────────────────────────────────────────────────────
# TASK 2: Multi-Input Prompt System
# ─────────────────────────────────────────────────────────────

# Multi-variable PromptTemplate
multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

def generate_multi_prompt(topic, audience, tone):
    """Generate a customized explanation prompt for given topic, audience, and tone."""
    return multi_input_template.format(topic=topic, audience=audience, tone=tone)

# ── All 3 required test cases ──
test_cases = [
    {"topic": "AI",            "audience": "beginners",  "tone": "friendly"},
    {"topic": "Python",        "audience": "kids",        "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers",   "tone": "technical"},
]

print("📌 Task 2 Output – Multi-Input Prompt System:")
print("-" * 55)
for idx, case in enumerate(test_cases, start=1):
    print(f"Test {idx}: {generate_multi_prompt(**case)}")
print("-" * 55)
````

## ✅ Task 3: Prompt Variations Engine (15%)

3 templates: **Teaching** | **Interview** | **Storytelling** — run for `topic = Machine Learning`

````python
# ─────────────────────────────────────────────────────────────
# TASK 3: Prompt Variations Engine
# ─────────────────────────────────────────────────────────────

# 3 separate PromptTemplates — one per style
variation_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step"
    ),
    "interview": PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 questions about {topic}"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a story"
    ),
}

def get_variation_prompt(style, topic):
    """Return a variation prompt based on the requested style."""
    if style not in variation_templates:
        raise ValueError(f"Unknown style '{style}'. Choose from: {list(variation_templates.keys())}")
    return variation_templates[style].format(topic=topic)

# ── Run all 3 styles for Machine Learning ──
topic = "Machine Learning"

print("📌 Task 3 Output – Prompt Variations Engine:")
print("-" * 55)
for style in variation_templates:
    print(f"[{style.upper()}]  →  {get_variation_prompt(style, topic)}")
print("-" * 55)
````

## ✅ Task 4: ChatPromptTemplate System (15%)

System + User message roles: **teacher** | **interviewer** | **motivator**

````python
# ─────────────────────────────────────────────────────────────
# TASK 4: ChatPromptTemplate System
# ─────────────────────────────────────────────────────────────

# Role-specific system behavior messages
role_system_messages = {
    "teacher":     "You are a patient and knowledgeable teacher who explains concepts clearly and step by step.",
    "interviewer": "You are a professional technical interviewer who asks insightful questions to assess understanding.",
    "motivator":   "You are an enthusiastic motivator who inspires learners by connecting topics to real-world impact.",
}

def build_chat_prompt(role, topic):
    """Build a ChatPromptTemplate dynamically based on role and topic."""
    if role not in role_system_messages:
        raise ValueError(f"Unknown role '{role}'. Choose from: {list(role_system_messages.keys())}")

    # System message — defines AI role/behavior
    system_template = SystemMessagePromptTemplate.from_template(role_system_messages[role])

    # Human message — dynamic user input
    human_template = HumanMessagePromptTemplate.from_template("Please help me understand {topic}.")

    # Assemble ChatPromptTemplate
    chat_prompt = ChatPromptTemplate.from_messages([system_template, human_template])

    return chat_prompt.format_messages(topic=topic)

# ── Run all 3 roles ──
print("📌 Task 4 Output – ChatPromptTemplate System:")
print("-" * 60)

test_combinations = [
    {"role": "teacher",     "topic": "Neural Networks"},
    {"role": "interviewer", "topic": "Deep Learning"},
    {"role": "motivator",   "topic": "Reinforcement Learning"},
]

for combo in test_combinations:
    messages = build_chat_prompt(**combo)
    print(f"\n🎭 Role: {combo['role'].upper()} | Topic: {combo['topic']}")
    for msg in messages:
        label = "[SYSTEM]" if msg.type == "system" else "[USER]"
        print(f"  {label} {msg.content}")

print("-" * 60)
````

## ✅ Task 5: Input Validation Layer (10%)

`validate_inputs(audience, tone)` — invalid inputs get default values instead of crashing.

````python
# ─────────────────────────────────────────────────────────────
# TASK 5: Input Validation Layer
# ─────────────────────────────────────────────────────────────

# Allowed values — defined as constants (not hardcoded inside logic)
VALID_AUDIENCES  = ["beginner", "intermediate", "expert"]
VALID_TONES      = ["formal", "casual", "fun"]
DEFAULT_AUDIENCE = "beginner"
DEFAULT_TONE     = "casual"

def validate_inputs(audience, tone):
    """
    Validate audience and tone inputs.
    Returns (audience, tone) with defaults assigned for any invalid values.
    """
    v_audience = audience.lower().strip()
    v_tone     = tone.lower().strip()

    if v_audience not in VALID_AUDIENCES:
        print(f"  ⚠️  Invalid audience '{audience}'. Defaulting to '{DEFAULT_AUDIENCE}'.")
        v_audience = DEFAULT_AUDIENCE

    if v_tone not in VALID_TONES:
        print(f"  ⚠️  Invalid tone '{tone}'. Defaulting to '{DEFAULT_TONE}'.")
        v_tone = DEFAULT_TONE

    return v_audience, v_tone

# ── Test cases — valid and invalid inputs ──
print("📌 Task 5 Output – Input Validation Layer:")
print("-" * 55)

test_inputs = [
    ("beginner",     "fun"),      # Both valid
    ("expert",       "formal"),   # Both valid
    ("student",      "casual"),   # Invalid audience
    ("intermediate", "humorous"), # Invalid tone
    ("novice",       "serious"),  # Both invalid
]

for aud, tn in test_inputs:
    print(f"\nInput  → audience='{aud}', tone='{tn}'")
    result = validate_inputs(aud, tn)
    print(f"Output → audience='{result[0]}', tone='{result[1]}'")

print("-" * 55)
````

## ✅ Task 6: Prompt Generator App (15%)

`generate_prompt(topic, audience, tone, style)` — full validation + template pipeline.

````python
# ─────────────────────────────────────────────────────────────
# TASK 6: Prompt Generator App
# ─────────────────────────────────────────────────────────────

VALID_STYLES  = ["teaching", "interview", "storytelling"]
DEFAULT_STYLE = "teaching"

# Style-specific PromptTemplates
generator_templates = {
    "teaching": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} teaching style"
    ),
    "interview": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Ask {audience}-level interview questions about {topic} in a {tone} tone"
    ),
    "storytelling": PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template="Explain {topic} for {audience} in a {tone} storytelling style"
    ),
}

def generate_prompt(topic, audience, tone, style):
    """
    Full prompt generation pipeline:
    Step 1 → Validate audience & tone
    Step 2 → Validate style
    Step 3 → Select PromptTemplate & return formatted prompt
    """
    # Step 1: Validate audience and tone
    v_audience, v_tone = validate_inputs(audience, tone)

    # Step 2: Validate style
    v_style = style.lower().strip()
    if v_style not in VALID_STYLES:
        print(f"  ⚠️  Invalid style '{style}'. Defaulting to '{DEFAULT_STYLE}'.")
        v_style = DEFAULT_STYLE

    # Step 3: Generate prompt using selected template
    return generator_templates[v_style].format(
        topic=topic, audience=v_audience, tone=v_tone
    )

# ── Test the Prompt Generator App ──
print("📌 Task 6 Output – Prompt Generator App:")
print("-" * 62)

app_tests = [
    {"topic": "Neural Networks",            "audience": "beginner",    "tone": "fun",    "style": "storytelling"},
    {"topic": "Natural Language Processing", "audience": "expert",      "tone": "formal", "style": "interview"},
    {"topic": "Python",                     "audience": "intermediate", "tone": "casual", "style": "teaching"},
    {"topic": "Computer Vision",            "audience": "student",      "tone": "chill",  "style": "quiz"},
]

for i, test in enumerate(app_tests, start=1):
    print(f"\n▶ Test {i}:")
    result = generate_prompt(**test)
    print(f"  ✅ {result}")

print("-" * 62)
````

## ✅ Task 7: Template Reusability Test (10%)

**ONE template → 5 different inputs → 5 different outputs.**

````python
# ─────────────────────────────────────────────────────────────
# TASK 7: Template Reusability Test
# ─────────────────────────────────────────────────────────────

# ONE reusable template — same structure every time
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 5 completely different input combinations
reusability_inputs = [
    {"topic": "Machine Learning",       "audience": "beginners",   "tone": "simple"},
    {"topic": "Transformers",            "audience": "researchers", "tone": "technical"},
    {"topic": "Data Preprocessing",     "audience": "analysts",    "tone": "practical"},
    {"topic": "Gradient Descent",       "audience": "students",    "tone": "visual"},
    {"topic": "Reinforcement Learning", "audience": "engineers",   "tone": "concise"},
]

print("📌 Task 7 Output – Template Reusability Test:")
print("🔁 ONE Template → 5 Different Outputs")
print("-" * 65)

for idx, inputs in enumerate(reusability_inputs, start=1):
    print(f"[Input {idx}]  {reusable_template.format(**inputs)}")

print("-" * 65)
print("✅ Same template structure | Different dynamic outputs | Fully reusable!")
````
