## Problem 1 — Smart Light Controller
Create a class to turn a smart light ON or OFF and display status.

```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def show_status(self):
        print(f"{self.name} is {self.status}")


light = SmartLight("Bedroom Light")
light.turn_on()
light.show_status()
```

## Problem 2 — Employee ID Card System
Store employee name, ID, and department and display details.

```python
class Employee:
    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


emp = Employee("Rahul", "EMP102", "IT")
emp.display()
```

## Problem 3 — Mobile Contact Record
Store contact name and phone number.

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c = Contact("Anita", "9876543210")
c.display()
```

## Problem 4 — Product Price Tag Generator
Store product name and price.

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display(self):
        print("Product:", self.name)
        print("Price: ₹", self.price)


p = Product("Headphones", 2499)
p.display()
```

## Problem 5 — Movie Rating Display System
Store movie name and rating.

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie:", self.name)
        print("Rating:", self.rating, "/ 5")


m = Movie("Inception", 4.8)
m.display()
```

## Problem 6 — Delivery Address Manager
Store customer name and address.

```python
class Delivery:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.name)
        print("Address:", self.address)


d = Delivery("Suman", "Hyderabad")
d.display()
```
