## ⚙️ Task 2: Advanced Preprocessing Function

The following function preprocesses raw text by:
- Removing URLs and email patterns
- Removing numbers
- Converting text to lowercase
- Handling repeated characters
- Removing special characters
- Removing extra spaces
- Removing short tokens (≤ 2 characters, except 'no' and 'not')
- Removing stopwords (except 'no' and 'not')

```python
import re

# Function to normalize repeated characters (soooo -> soo)
def normalize_repeated_chars(word):
    return re.sub(r'(.)\1{2,}', r'\1\1', word)

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""

    # Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Convert to lowercase
    text = text.lower()

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Handle repeated characters per word
    tokens = [normalize_repeated_chars(word) for word in tokens]

    # Remove very short tokens (≤2 except 'no', 'not')
    tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    clean_sentence = " ".join(tokens)

    return tokens, clean_sentence


# TEST CASES (required)
print(preprocess_text("I have 2 dogs"))
print(preprocess_text("This is good"))
print(preprocess_text("WOW!!! This is GREAT!!!"))
print(preprocess_text("soooo goooood!!!"))
print(preprocess_text("Visit http://example.com now"))
```

```python
def display_results(text_list):
    for text in text_list:
        tokens, clean = preprocess_text(text)

        print("Original Text:", text)
        print("Cleaned Tokens:", tokens)
        print("Cleaned Sentence:", clean)
        print("-" * 50)

```

```python
from collections import Counter

def analyze_text(text_list):
    all_tokens = []

    # Collect all tokens
    for text in text_list:
        tokens, _ = preprocess_text(text)
        all_tokens.extend(tokens)

    # Frequency analysis
    counter = Counter(all_tokens)

    print("Top 5 words:", counter.most_common(5))
    print("Least 5 words:", counter.most_common()[-5:])
    print("-" * 50)

    # Sentence-wise analysis
    for text in text_list:
        tokens, clean = preprocess_text(text)

        total = len(tokens)
        unique = len(set(tokens))
        avg_len = round(sum(len(t) for t in tokens) / total, 2) if total > 0 else 0

        print("Sentence:", clean)
        print("Total:", total, "| Unique:", unique, "| Avg Length:", avg_len)
        print("-" * 50)
```

```python
analyze_text(test_sentences)
```

**Which sentence retained the most meaningful tokens after cleaning?**

```python
from collections import Counter

def frequency_analysis(text_list):
    all_tokens = []

    # Collect tokens from all sentences
    for text in text_list:
        tokens, _ = preprocess_text(text)
        all_tokens.extend(tokens)

    counter = Counter(all_tokens)

    print("Top 10 most frequent words:")
    print(counter.most_common(10))

    print("\nTop 5 least frequent words:")
    print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print("Tokens:", result["tokens"])
print("\nClean Sentences:", result["clean_sentences"])
```

## ⚠️ Task 7: Error Handling

The preprocessing function successfully handles edge cases:

- Empty string → returns empty tokens and sentence
- Only emojis → removed during cleaning, resulting in empty output
- Only numbers → removed during preprocessing, resulting in empty output

This ensures the pipeline is robust and does not break on invalid or noisy inputs.

```python
edge_cases = [
    "",                 # Empty string
    "😍😍🔥🔥",         # Only emojis
    "1234567890"        # Only numbers
]

for case in edge_cases:
    tokens, clean = preprocess_text(case)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    print("-" * 40)
```
